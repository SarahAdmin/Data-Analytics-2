```python
import pandas as pd
```

```python
df = pd.read_csv('ticket_data.csv',encoding='latin-1')
```

```python
Attendees_tickets = df.groupby('Attendee first name')['Ticket price'].sum()
```

```python
quantity_count = df['COUNTIF'].value_counts()
```

```python
ticket_count = df['Ticket price'].sum()
```

```python
ticket_count
```

```python
import matplotlib.pyplot as plt
```

```python
plt.barh(quantity_count,ticket_count, color='rebeccapurple',alpha=0.5)
plt.xlabel('Ticket Price')
plt.ylabel('Quantity')
plt.title('Ticket Prices')
plt.show()
```
