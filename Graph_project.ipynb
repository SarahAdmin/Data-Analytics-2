```python
import pandas as pd
```

```python
data = {'Name':['Lee Cox','Lemon Jelly','Matt Bond','Pamela Adams'],'Salary':[35000,40000,35000,70000]}
df = pd.DataFrame(data)
```

```python
x = df['Name']
y = df['Salary']
```

```python
plt.bar(x,y,color='orange')
plt.title('Kings Employee Salary')
plt.ylabel('Salary')
```
