```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Videoshoot data.xlsx',sheet_name='Venues for Video Shoot')
```

```python
df
```

```python
plt.barh(df['Property Name'],df['Price for 2/3 Hours'],color='blue')
plt.xlabel('Price for 2/3 Hours')
plt.ylabel('Property Name')
plt.title('Property Prices')
plt.show()
```

```python
plt.plot(df['Price for 2/3 Hours'],df['Property Name'],color='orange',alpha=0.5)
plt.xlabel('Price for 2/3 Hours')
plt.ylabel('Property Name')
plt.title('Property Prices')
plt.show()
```

```python
plt.scatter(df['Price for 2/3 Hours'],df['Property Name'],color='rebeccapurple',marker='s',alpha=0.5)
plt.xlabel('Price for 2/3 Hours')
plt.ylabel('Property Name')
plt.title('Property Prices')
plt.axhline(color='black',alpha=0.5,linestyle='-')
plt.axvline(color='black',alpha=0.5,linestyle='-')
plt.show()
```

```python
df1 = pd.read_csv('theory_test_data.csv',encoding='latin-1')
```

```python
plt.barh(df1['Name'],df1['Score_out_of_50'],color='orange',alpha=0.5)
plt.xlabel('Score')
plt.ylabel('Name')
plt.title('Theory Test Results')
plt.show()
```

```python
plt.plot(df1['Score_out_of_50'],df1['Name'],alpha=0.5,color='rebeccapurple',marker='o')
plt.xlabel('Score')
plt.ylabel('Name')
plt.title('Theory Test Results')
plt.axhline(color='black',alpha=0.5,linestyle='--')
plt.axvline(color='black',alpha=0.5,linestyle='--')
plt.show()
```
